```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('hotel_bookings.csv')
```

```python
### Drop duplicate rows
df.drop_duplicates(inplace=True)
df.shape
```

```python
# missing values
df.isnull().sum()
```

```python
for col in df.columns:
    if np.sum(df[col].isnull())>(df.shape[0]*0.8):
        df.drop(columns=col, inplace=True)
print(df.shape)
```

```python
df['country'].mode(),df['children'].mean(),df['agent'].median()
```

```python
def impute_mean(data):
    fill_mean = data.fillna(data.mean())
    return fill_mean
```

```python
def impute_mode(data):
    fill_mode = data.fillna(str(data.mode())) #convert the column to string in order to fill with the mode
    return fill_mode
```

```python
def impute_median(data):
    fill_median = data.fillna(data.median())
    return fill_median
```

```python
df['children'] = impute_mean(df["children"])
```

```python
df['country'] = impute_mode(df["country"])
```

```python
df['agent'] = impute_median(df["agent"])
```

```python
# merging the year, month and day then change the type to datetime format
df['arrival_date'] = pd.to_datetime(df.arrival_date_year.astype(str)+'/'+df.arrival_date_month.astype(str)+'/'+df.arrival_date_day_of_month.astype(str))
```

```python
# changing the datatype of the children and agent
df[["children","agent",]]=df[["children","agent"]].astype('int64')
```
